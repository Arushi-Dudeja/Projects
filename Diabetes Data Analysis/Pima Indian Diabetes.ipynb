#### Step 3: Exploratory Data Analysis

```python
sns.displot(pima['BloodPressure'],kde=False,bins=6)
plt.show()
```

Just by looking at the 'BloodPressure' variable we can say that the most number of observations are found to be in the range of 60 to 80. We can therefore say that the mode of the variable 'BloodPressure' should lie somewhere between 60 and 80.

#### Question: What is the 'BMI' for the person having the highest 'Glucose'? Also, mention what is the highest glucose?

```python
pima[pima['Glucose']==pima['Glucose'].max()]['BMI']
```

The 'BMI' of the person having the maximum 'Glucose' is 42.9. The value of the highest 'Glucose' level is 199.

#### Question: What is the mean of the variable 'BMI'? What is the median of the variable 'BMI'? What is the mode of the variable 'BMI'? Are the three measures of central tendency equal?

```python
m1 = pima['BMI'].mean()
print(m1)
m2 = pima['BMI'].median()
print(m2)
m3= pima['BMI'].mode().values
print(m3)
```

The three measures of Central Tendency of the variable 'BMI' is very similar but not exactly equal.

#### Question: How many entries (people) are above the mean level of 'Glucose'?

```python
pima[pima['Glucose']>pima['Glucose'].mean()].shape[0]
```

#### Question: How many entries (people) have their 'BloodPressure' equal to the median of 'BloodPressure' and their 'BMI' less than the median of 'BMI'?

```python
pima[(pima['BloodPressure']==pima['BloodPressure'].median()) & (pima['BMI']<pima['BMI'].median())].shape[0]
```

#### Question: Below are the scatterplots of 'Glucose', 'SkinThickness' and 'DiabetesPedigreeFunction' against each other. Explain each of the plots

```python
sns.pairplot(data=pima,vars=['Glucose', 'SkinThickness', 'DiabetesPedigreeFunction'],hue='Outcome')
plt.show()
```

Let us take two variables at a time and explain the plots.
    
* 'Glucose' vs 'Skin Thickness' - We notice that as the value of 'Glucose' increase the value of 'SkinThickness' does not increase that much. It remains more or less constant. On top of that we see that the people who have diabetes have a higher 'Glucose' level but the 'Skinthickness' value for the diabetic as well the non-diabetic people remains almost constant in this graph.
    
* 'Glucose' vs 'DiabetesPedigreeFunction' - We notice that the there is a very slight increase in the 'DiabetesPedigreeFunction' with the increase in 'Glucose' values. Also, people who are diabetic has higher 'Glucose' values. Along with that we see that there are few data points which indicates that the 'DiabetesPedigreeFunction' has increased for higher values 'Glucose'.
    
* 'Glucose' vs 'SkinThickness' - The values for 'Glucose' and 'SkinThickness' also remains almost constant and there is not huge change or difference in the values given the fact that whether people are diabetic or not.

#### Question: Plot the scatterplot between 'Glucose' and 'Insulin'. What does the plot explain?

```python
plt.scatter(x='Glucose',y='Insulin',data=pima)
plt.show()
```

We see that as the value of 'Glucose' increases there is a clear trend or pattern in the increase of the value of 'Insulin'.

#### Question: Plot a histogram for variable Age to understand the number of women in different Age groups given that they have and do not have diabetes. Explain and compare the two graphs.

```python
plt.hist(pima[pima['Outcome']==1]['Age'], bins =5)
plt.title('Distribution of Age for Women who has Diabetes')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(pima[pima['Outcome']==0]['Age'], bins =5)
plt.title('Distribution of Age for Women who do not have Diabetes')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

We see that the women of almost all age groups have diabetes with the age group of 20 to 30 recording the most number of cases. On the other hand, we see that mostly younger women do not have diabetes.

#### Question: Plot the correlation matrix in the form of a heatmap. Explain the heatmap.

```python
corr_matrix = pima.iloc[:,0:8].corr()
corr_matrix
```

```python
sns.heatmap(corr_matrix, annot = True)

# display the plot
plt.show()
```

In this correlation matrix, we have not included the information about whether the person is diabetic or not. We see that there is a positive correlation of 0.54 between 'Pregnancies' and 'Age. This means that the number of 'Pregnancies' linearly increases with 'Age'. We have to dig deep over and check the correlation matrix taking the information of diabetes into account. Likewise, we have to check and infer from the correlation between all of the variables taken two at a time.
